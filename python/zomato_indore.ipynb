```python
tempdfx = tempdfx.drop(["cuisines"], axis="columns")
```

```python
tempdfx.head()
```

```python
dfdummy = pd.get_dummies(tempdfx['rating_text'])
dfdummy.head()
```

```python
mergeddf = pd.concat([tempdfx,dfdummy], axis="columns")
mergeddf
```

```python
dfx = mergeddf.drop(["rating_text","aggregate_rating"], axis="columns")
```

```python
dfy = mergeddf["aggregate_rating"]
dfy = dfy.astype(int)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dfx,dfy, test_size=0.3)
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
